# Explore Image Processing
This notebook demonstrates how the sample image is converted to a 1-bit format for printing.

```python
from pathlib import Path
from PIL import Image, ImageOps
from imaging.process import to_1bit
```

```python
image_path = Path('../20250902_0027_Party in the Park_simple_compose_01k43p7hzyfcst3ff4nfs6hatn.png')
original = Image.open(image_path)
original
```

```python
# Convert to grayscale
img_gray = ImageOps.exif_transpose(original).convert('L')
img_gray
```

```python
# Resize to printer width
printer_width = 463
ratio = printer_width / img_gray.width
new_height = int(img_gray.height * ratio)
img_resized = img_gray.resize((printer_width, new_height), Image.LANCZOS)
img_resized
```

```python
# Convert to 1-bit
canvas = Image.new('L', (printer_width, new_height), 255)
canvas.paste(img_resized, (0,0))
final_img = canvas.convert('1')
final_img
```

```python
# Using the helper function directly
final_img_direct = to_1bit(image_path.read_bytes(), printer_width)
final_img_direct
```
